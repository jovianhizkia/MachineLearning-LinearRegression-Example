##A. Variable splitting to dependent and independent variable

```python
# Memisahkan variabel bebas (independen)
dfx=df.iloc[:, 1:-1]

dfx
```

```python
# Memisahkan variabel terikat (dependen)
dfy=df.iloc[:, -1:]

dfy
```

##C. Data pre-processing

**Proses ini dilakukan untuk menghilangkan data yang kosong atau NULL pada dataframe (tabel). Kemudian, jika terdapat data yang kosong, data tersebut akan diisi menggunakan metode fillna() atau ffill().**

```python
# Menampilkan data yang kosong di variabel x
dfx.isna()
```

```python
# Menghitung jumlah data yang kosong di setiap kolom x
dfx.isna().sum()
```

```python
# Menampilkan data yang kosong di variabel y
dfy.isna()
```

```python
# Menghitung jumlah data yang kosong di setiap kolom y
dfy.isna().sum()
```

```python
# Menghilangkan data kosong pada variabel x
dfx.dropna(inplace=True)

dfx.isna().sum()
```

```python
# Menghilangkan data kosong pada variabel y
dfy.dropna(inplace=True)

dfy.isna().sum()
```

```python
# Mengisi data yang kosong pada variabel x dengan metode ffill
dfx=dfx.ffill()
dfx
```

```python
# Menggantikan data yang kosong pada variabel y dengan metode ffill
dfy=dfy.ffill()
dfy
```

##D. Training and testing data splitting

```python
from sklearn.model_selection import train_test_split

# Melakukan splitting pada dataset x dan y menjadi data train dan data test
xTrain, xTest, yTrain, yTest=train_test_split(dfx, dfy, test_size=0.25, random_state=0)

# Menampilkan data train dan data test
print("xTrain shape:", xTrain.shape)
print("yTrain shape:", yTrain.shape)
print("xTest shape:", xTest.shape)
print("yTest shape:", yTest.shape)
```

##E. Linear regression

###Prediction using test data

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Membuat fungsi regresi linear
linreg=LinearRegression()

# Menyesuaikan model dengan data train
linreg.fit(xTrain, yTrain)

# Memprediksi data Y baru berdasarkan data test
yPred=linreg.predict(xTest)

# Menampilkan intercept dari regresi
linreg.intercept_
```

```python
# Menampilkan koefisien dari regresi
linreg.coef_
```

###Comparing the predicted data and actual data

```python
# Menampilkan nilai y hasil prediksi dan nilai y yang sebenarnya
print("Nilai prediksi:", yPred, "\n\n", "Nilai sebenarnya:", yTest.values)
```

##F. Data evaluation

###Data normalization

```python
from sklearn.preprocessing import StandardScaler

# Membuat fungsi normalisasi data
normalize=StandardScaler()

# Melakukan normalization pada data train x dan data test x
xTrain, xTest=normalize.fit_transform(xTrain), normalize.transform(xTest)

# Membuat tabel dataframe baru berdasarkan data train dan test yang sudah dinormalisasi
dfxTrain=pd.DataFrame(xTrain, columns=dfx.columns)
dfxTest=pd.DataFrame(xTest, columns=dfx.columns)
```
